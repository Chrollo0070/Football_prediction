```python
# ===== Football Match Outcome Modeling: Enhanced Insights =====

import pandas as pd
import numpy as np
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, confusion_matrix,
    mean_absolute_error, mean_squared_error
)
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Load data
try:
    df_fifa = pd.read_csv('fifa2020-2024.csv')
    df_results = pd.read_csv('result1.csv')
except FileNotFoundError:
    print("Error: Ensure 'fifa2020-2024.csv' and 'result1.csv' are in the same directory.")
    raise

print("FIFA Rankings DataFrame (first 5 rows):")
display(df_fifa.head())
print("\nFIFA Rankings DataFrame Info:")
df_fifa.info()
print(f"\nFIFA Rankings DataFrame has {df_fifa.shape[0]} rows and {df_fifa.shape[1]} columns.")

print("\nMatch Results DataFrame (first 5 rows):")
display(df_results.head())
print("\nMatch Results DataFrame Info:")
df_results.info()
print(f"\nMatch Results DataFrame has {df_results.shape[0]} rows and {df_results.shape[1]} columns.")

# 2) Preprocessing and robust merging (pre-rename to avoid suffix confusion)
df_fifa['date'] = pd.to_datetime(df_fifa['date'])
df_results['date'] = pd.to_datetime(df_results['date'])

df_fifa = df_fifa.sort_values('date').reset_index(drop=True)
df_results = df_results.sort_values('date').reset_index(drop=True)

# Support either 'total.points' or 'total_points' column names
if 'total.points' in df_fifa.columns:
    fifa_points_col = 'total.points'
elif 'total_points' in df_fifa.columns:
    fifa_points_col = 'total_points'
else:
    raise KeyError("FIFA data is missing 'total.points' or 'total_points' column.")

# Build disjoint schemas for home and away to avoid ambiguous suffixes
fifa_home = df_fifa.rename(columns={'rank': 'home_rank', fifa_points_col: 'home_points'})[['date', 'team', 'home_rank', 'home_points']]
fifa_away = df_fifa.rename(columns={'rank': 'away_rank', fifa_points_col: 'away_points'})[['date', 'team', 'away_rank', 'away_points']]

# Home merge
df_merged = pd.merge_asof(
    df_results, fifa_home,
    left_on='date', right_on='date',
    left_by='home_team', right_by='team',
    direction='backward'
).rename(columns={'team': 'home_team_fifa_name'})

# Away merge
df_merged = pd.merge_asof(
    df_merged, fifa_away,
    left_on='date', right_on='date',
    left_by='away_team', right_by='team',
    direction='backward'
).rename(columns={'team': 'away_team_fifa_name'})

# Ensure FIFA columns exist, then drop rows missing them
required_fifa_cols = ['home_rank', 'home_points', 'away_rank', 'away_points']
missing_required = [c for c in required_fifa_cols if c not in df_merged.columns]
if missing_required:
    raise KeyError(f"Missing expected columns after merge: {missing_required}")
df_merged.dropna(subset=required_fifa_cols, inplace=True)
print(f"\nAfter dropping rows with missing FIFA rankings, df_merged has {df_merged.shape[0]} rows.")

# 3) Feature engineering
df_merged['rank_difference'] = df_merged['home_rank'] - df_merged['away_rank']
df_merged['points_difference'] = df_merged['home_points'] - df_merged['away_points']
df_merged['home_win'] = (df_merged['home_score'] > df_merged['away_score']).astype(int)
df_merged['score_difference'] = df_merged['home_score'] - df_merged['away_score']

# One-hot encode tournament if present
if 'tournament' in df_merged.columns:
    df_merged = pd.get_dummies(df_merged, columns=['tournament'], prefix='tournament')
else:
    print("Warning: 'tournament' column not found; proceeding without tournament dummies.")

# Ensure 'neutral' exists and is integer
if 'neutral' not in df_merged.columns:
    df_merged['neutral'] = 0
df_merged['neutral'] = df_merged['neutral'].astype(int)

print("\nMerged DataFrame (first 5 rows after feature engineering):")
display(df_merged.head())
print("\nMerged DataFrame Info:")
df_merged.info()

# 4) Build feature matrix and targets
engineered_features = ['rank_difference', 'points_difference', 'neutral']
tournament_features = [c for c in df_merged.columns if c.startswith('tournament_')]
features = [f for f in engineered_features + tournament_features if f in df_merged.columns]

X = df_merged[features]
y_clf = df_merged['home_win']
y_reg = df_merged['score_difference']

# Single, consistent split for all targets
X_train, X_test, y_train_clf, y_test_clf, y_train_reg, y_test_reg = train_test_split(
    X, y_clf, y_reg, test_size=0.2, random_state=42
)

# 5) Scale using the same scaler for both tasks
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test_scaled  = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

# Safety: impute any residual NaNs post-scaling (should be rare)
X_train_scaled = X_train_scaled.fillna(X_train_scaled.mean())
X_test_scaled  = X_test_scaled.fillna(X_test_scaled.mean())

print("\nClassification task shapes:")
print("X_train_scaled:", X_train_scaled.shape)
print("X_test_scaled:", X_test_scaled.shape)
print("y_train_clf:", y_train_clf.shape)
print("y_test_clf:", y_test_clf.shape)

print("\nRegression task shapes:")
print("X_train_scaled:", X_train_scaled.shape)
print("X_test_scaled:", X_test_scaled.shape)
print("y_train_reg:", y_train_reg.shape)
print("y_test_reg:", y_test_reg.shape)

# 6) Train models
model_clf = LogisticRegression(random_state=42, max_iter=1000)
model_clf.fit(X_train_scaled, y_train_clf)

model_reg = RandomForestRegressor(random_state=42)
model_reg.fit(X_train_scaled, y_train_reg)

print("\nModels trained successfully.")

# 7) Evaluate models
# Classification
y_pred_clf = model_clf.predict(X_test_scaled)
accuracy = accuracy_score(y_test_clf, y_pred_clf)
precision = precision_score(y_test_clf, y_pred_clf)
recall = recall_score(y_test_clf, y_pred_clf)
conf_matrix = confusion_matrix(y_test_clf, y_pred_clf)

print("\n--- Classification Model Evaluation ---")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print("Confusion Matrix:")
print(conf_matrix)

# Regression
y_pred_reg = model_reg.predict(X_test_scaled)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)

print("\n--- Regression Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")

# 8) Helpers: richer team insights

def _last_n_matches(df, team, n=5):
    m = df[(df['home_team'] == team) | (df['away_team'] == team)].copy()
    if m.empty:
        return m
    m = m.sort_values('date', ascending=False).head(n).copy()
    def opponent(row):
        return row['away_team'] if row['home_team'] == team else row['home_team']
    def venue(row):
        if row.get('neutral', 0) == 1:
            return 'Neutral'
        return 'Home' if row['home_team'] == team else 'Away'
    def result(row):
        hs, as_ = row['home_score'], row['away_score']
        if row['home_team'] == team:
            if hs > as_:
                return 'W'
            elif hs == as_:
                return 'D'
            return 'L'
        else:
            if as_ > hs:
                return 'W'
            elif as_ == hs:
                return 'D'
            return 'L'
    def scoreline(row):
        if row['home_team'] == team:
            return f"{row['home_score']}-{row['away_score']}"
        return f"{row['away_score']}-{row['home_score']}"
    m['Opponent'] = m.apply(opponent, axis=1)
    m['Venue'] = m.apply(venue, axis=1)
    m['Result'] = m.apply(result, axis=1)
    m['Scoreline'] = m.apply(scoreline, axis=1)
    cols = ['date', 'Opponent', 'Venue', 'tournament', 'Result', 'Scoreline', 'home_team', 'away_team', 'home_score', 'away_score']
    cols = [c for c in cols if c in m.columns]
    m = m[cols].rename(columns={'date': 'Date', 'tournament': 'Tournament'})
    return m

def _form_summary(last_df, team):
    if last_df.empty:
        return {
            'form_string': '-',
            'W': 0, 'D': 0, 'L': 0,
            'GF': 0, 'GA': 0, 'GD': 0,
            'Win%': 0.0
        }
    res = last_df['Result'].tolist()
    W = res.count('W')
    D = res.count('D')
    L = res.count('L')
    # Goals for/against from perspective of team
    def gf(row):
        return row['home_score'] if row['home_team'] == team else row['away_score']
    def ga(row):
        return row['away_score'] if row['home_team'] == team else row['home_score']
    GF = last_df.apply(gf, axis=1).sum()
    GA = last_df.apply(ga, axis=1).sum()
    GD = GF - GA
    n = len(last_df)
    win_pct = round(100.0 * W / n, 1)
    return {
        'form_string': '-'.join(res),
        'W': W, 'D': D, 'L': L,
        'GF': int(GF), 'GA': int(GA), 'GD': int(GD),
        'Win%': win_pct
    }

def _current_fifa_snapshot(team):
    tdf = df_fifa[df_fifa['team'] == team]
    if tdf.empty:
        return None
    latest = tdf.sort_values('date', ascending=False).iloc[0]
    pts_col = 'total.points' if 'total.points' in latest.index else ('total_points' if 'total_points' in latest.index else None)
    points = float(latest[pts_col]) if pts_col else None
    return {'rank': int(latest['rank']), 'points': points, 'date': latest['date']}

def _head_to_head(df, team_a, team_b, n=5):
    h2h = df[((df['home_team'] == team_a) & (df['away_team'] == team_b)) |
             ((df['home_team'] == team_b) & (df['away_team'] == team_a))].copy()
    if h2h.empty:
        return h2h, {'A_wins': 0, 'B_wins': 0, 'Draws': 0}
    h2h = h2h.sort_values('date', ascending=False)
    def res_for(team, row):
        hs, as_ = row['home_score'], row['away_score']
        if row['home_team'] == team:
            return 'W' if hs > as_ else ('D' if hs == as_ else 'L')
        else:
            return 'W' if as_ > hs else ('D' if as_ == hs else 'L')
    # Summary over all H2H
    A_wins = sum(res_for(team_a, r) == 'W' for _, r in h2h.iterrows())
    B_wins = sum(res_for(team_b, r) == 'W' for _, r in h2h.iterrows())
    Draws = sum(res_for(team_a, r) == 'D' for _, r in h2h.iterrows())
    # Last n table
    show = h2h.head(n).copy()
    def opp(row, team):
        return row['away_team'] if row['home_team'] == team else row['home_team']
    def venue_for(team, row):
        if row.get('neutral', 0) == 1:
            return 'Neutral'
        return 'Home' if row['home_team'] == team else 'Away'
    # Build compact table from the perspective of team_a
    show['Opponent'] = show.apply(lambda r: opp(r, team_a), axis=1)
    show['Venue(A)'] = show.apply(lambda r: venue_for(team_a, r), axis=1)
    show['Result(A)'] = show.apply(lambda r: res_for(team_a, r), axis=1)
    def scoreline_from_a(row):
        if row['home_team'] == team_a:
            return f"{row['home_score']}-{row['away_score']}"
        return f"{row['away_score']}-{row['home_score']}"
    show['Scoreline(A)'] = show.apply(scoreline_from_a, axis=1)
    keep_cols = ['date', 'Opponent', 'Venue(A)', 'tournament', 'Result(A)', 'Scoreline(A)']
    keep_cols = [c for c in keep_cols if c in show.columns]
    show = show[keep_cols].rename(columns={'date': 'Date', 'tournament': 'Tournament'})
    return show, {'A_wins': int(A_wins), 'B_wins': int(B_wins), 'Draws': int(Draws)}

# 9) Prediction function with insights
feature_columns = X_train.columns  # reference for ordering

def predict_match_outcome(home_team, away_team, tournament, neutral):
    """
    Predict winner and scoreline, and print meaningful team insights (last 5, form, H2H).
    Returns (winner_str, scoreline_str) or (None, None) if team ranks not found.
    """
    print(f"\n=== Preview: {home_team} vs {away_team} ===")

    # FIFA snapshots
    home_fifa = _current_fifa_snapshot(home_team)
    away_fifa = _current_fifa_snapshot(away_team)
    if home_fifa is None:
        print(f"Error: FIFA ranking data not found for {home_team}.")
        return None, None
    if away_fifa is None:
        print(f"Error: FIFA ranking data not found for {away_team}.")
        return None, None

    print("\n-- Current FIFA Snapshot --")
    print(f"{home_team}: Rank {home_fifa['rank']}, Points {home_fifa['points']:.2f} (as of {home_fifa['date'].date()})")
    print(f"{away_team}: Rank {away_fifa['rank']}, Points {away_fifa['points']:.2f} (as of {away_fifa['date'].date()})")

    # Last 5 for both teams
    last5_home = _last_n_matches(df_results, home_team, n=5)
    last5_away = _last_n_matches(df_results, away_team, n=5)

    print(f"\n-- {home_team}: Last 5 Matches --")
    if not last5_home.empty:
        display(last5_home)
        hs = _form_summary(last5_home, home_team)
        print(f"Form: {hs['form_string']} | W-D-L: {hs['W']}-{hs['D']}-{hs['L']} | GF/GA/GD: {hs['GF']}/{hs['GA']}/{hs['GD']} | Win%: {hs['Win%']}%")
    else:
        print("No recent matches found.")

    print(f"\n-- {away_team}: Last 5 Matches --")
    if not last5_away.empty:
        display(last5_away)
        as_ = _form_summary(last5_away, away_team)
        print(f"Form: {as_['form_string']} | W-D-L: {as_['W']}-{as_['D']}-{as_['L']} | GF/GA/GD: {as_['GF']}/{as_['GA']}/{as_['GD']} | Win%: {as_['Win%']}%")
    else:
        print("No recent matches found.")

    # Head-to-head
    h2h_table, h2h_sum = _head_to_head(df_results, home_team, away_team, n=5)
    print(f"\n-- Head-to-Head (last {min(5, len(h2h_table)) if not h2h_table.empty else 0}) --")
    if not h2h_table.empty:
        display(h2h_table)
        print(f"H2H Summary: {home_team} wins: {h2h_sum['A_wins']} | Draws: {h2h_sum['Draws']} | {away_team} wins: {h2h_sum['B_wins']}")
    else:
        print("No recent head-to-head matches found.")

    # Build model input
    rank_diff = home_fifa['rank'] - away_fifa['rank']
    points_diff = home_fifa['points'] - away_fifa['points']
    input_data = {
        'rank_difference': [rank_diff],
        'points_difference': [points_diff],
        'neutral': [int(bool(neutral))]
    }
    input_df = pd.DataFrame(input_data)

    # Tournament dummies as in training
    for col in feature_columns:
        if col.startswith('tournament_') and col not in input_df.columns:
            input_df[col] = 0
    specific_tournament_col = f'tournament_{tournament}'
    if specific_tournament_col in feature_columns:
        input_df[specific_tournament_col] = 1
    else:
        print(f"Warning: Tournament '{tournament}' not seen during training; treating as other/zero vector.")

    # Align columns and scale
    for col in feature_columns:
        if col not in input_df.columns:
            input_df[col] = 0
    input_df = input_df[feature_columns]
    input_scaled = pd.DataFrame(scaler.transform(input_df), columns=input_df.columns, index=input_df.index)

    # Predict
    pred_clf = model_clf.predict(input_scaled)[0]
    proba_clf = model_clf.predict_proba(input_scaled)[0]
    pred_reg = model_reg.predict(input_scaled)[0]

    predicted_score_difference = int(round(pred_reg))
    winner, home_score, away_score = "", 0, 0

    if pred_clf == 1:
        winner = f"{home_team} Wins"
        score_diff = max(1, predicted_score_difference)
        if score_diff == 1:
            home_score, away_score = 2, 1
        elif score_diff == 2:
            home_score, away_score = 3, 1
        elif score_diff > 2:
            home_score, away_score = score_diff + 1, 1
        else:
            home_score, away_score = 1, 0
    else:
        if predicted_score_difference < 0:
            winner = f"{away_team} Wins"
            score_diff = max(1, abs(predicted_score_difference))
            if score_diff == 1:
                away_score, home_score = 2, 1
            elif score_diff == 2:
                away_score, home_score = 3, 1
            elif score_diff > 2:
                away_score, home_score = score_diff + 1, 1
            else:
                away_score, home_score = 1, 0
        else:
            winner = "Draw"
            home_score, away_score = 1, 1

    print(f"\n-- Model Outlook --")
    print(f"Home Win Probability: {proba_clf[1]:.2f} | Not Home Win: {proba_clf[0]:.2f}")
    print(f"Predicted Score Difference (regression): {pred_reg:.2f}")
    scoreline = f"{home_score}-{away_score}"
    print(f"Prediction: {winner} | Suggested Scoreline: {scoreline}")
    return winner, scoreline

# 10) Visualizations (optional)
if hasattr(model_reg, 'feature_importances_'):
    feature_importances_reg = model_reg.feature_importances_
    fi_series = pd.Series(feature_importances_reg, index=X_train_scaled.columns).sort_values(ascending=False)
    plt.figure(figsize=(12, 7))
    sns.barplot(x=fi_series.head(15).values, y=fi_series.head(15).index, palette='viridis')
    plt.title('Top 15 Feature Importances (Random Forest Regressor)', fontsize=16)
    plt.xlabel('Importance (Relative)', fontsize=12)
    plt.ylabel('Feature', fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print("Regression model does not expose 'feature_importances_'.")

if hasattr(model_clf, 'coef_'):
    coef = model_clf.coef_[0]
    coef_series = pd.Series(coef, index=X_train_scaled.columns).sort_values(key=np.abs, ascending=False)
    plt.figure(figsize=(12, 7))
    sns.barplot(x=coef_series.head(15).values, y=coef_series.head(15).index, palette='coolwarm')
    plt.title('Top 15 Most Influential Coefficients (Logistic Regression)', fontsize=16)
    plt.xlabel('Coefficient Value', fontsize=12)
    plt.ylabel('Feature', fontsize=12)
    plt.tight_layout()
    plt.show()
else:
    print("Classification model does not expose 'coef_'.")

plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
    xticklabels=['Predicted No Home Win', 'Predicted Home Win'],
    yticklabels=['Actual No Home Win', 'Actual Home Win']
)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.title('Confusion Matrix (Logistic Regression Classifier)', fontsize=16)
plt.show()

# 11) Example usage
home_team_example = "Portugal"
away_team_example = "Spain"
tournament_example = "Friendly"
neutral_example = False

predicted_winner, predicted_scoreline = predict_match_outcome(
    home_team_example, away_team_example, tournament_example, neutral_example
)
if predicted_winner and predicted_scoreline:
    print(f"\n--- Final Prediction ---")
    print(f"Predicted Outcome: {predicted_winner}")
    print(f"Predicted Scoreline: {predicted_scoreline}")
```
